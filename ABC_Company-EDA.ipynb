```python

# Distribution of employees across teams
team_distribution = df['Team'].value_counts()
team_percentage = (team_distribution / len(df)) * 100

print(team_distribution)
print(team_percentage)
```

```python
plt.figure(figsize=(10, 6))
team_percentage.plot(kind='bar', color='skyblue')
plt.title('Percentage Distribution of Employees by Team')
plt.ylabel('Percentage (%)')
plt.xlabel('Teams')
plt.show()
```

Insight:
The New Orleans Pelicans lead with the highest percentage of employees, suggesting a larger team size relative to others.
The Minnesota Timberwolves and Orlando Magic are at the lower end with the smallest percentage of employees.

```python
# 2 Distribution of employees by position
position_distribution = df['Position'].value_counts()
print(position_distribution)
```

```python
# Horizontal Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x=position_distribution.values, y=position_distribution.index, palette='viridis')
plt.title('Distribution of Employees by Position')
plt.xlabel('Number of Employees')
plt.ylabel('Position')
plt.show()
```

```python
# Age group distribution
age_bins = [20, 25, 30, 35, 40, 50]
age_labels = ['20-25', '25-30', '30-35', '35-40', '40+']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

age_group_distribution = df['Age Group'].value_counts()
print(age_group_distribution)
```

```python
# Define age bins and labels
bins = [15, 20, 25, 30, 35, 40]
labels = ['15-20', '21-25', '26-30', '31-35', '36-40']

# Create a new column 'Age Group'
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(8, 6))
sns.histplot(df['Age Group'], discrete=True)

# Adding labels and title
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
salary_by_team_position = df.groupby(['Team', 'Position'])['Salary'].sum()
highest_salary_team_position = salary_by_team_position.idxmax()
print(highest_salary_team_position)
print(salary_by_team_position.max())
```

```python
df1 = df.pivot_table(index='Position', columns='Team', values='Salary')
plt.figure(figsize=(8, 6))
sns.heatmap(df1, annot=True, fmt=".0f", cmap='YlGnBu', linewidths=.5)
plt.title('Team and Position Salary Expenditure')
plt.xlabel('Team')
plt.ylabel('Position')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation between age and salary
correlation = df['Age'].corr(df['Salary'])
print(f'Correlation between Age and Salary: {correlation}')

# Visualize with a scatter plot
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Age vs. Salary')
plt.show()
```

```python
# Plot a pie chart
import pandas as pd
import matplotlib.pyplot as plt
position_distribution = df['Position'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(position_distribution, labels=position_distribution.index, autopct='%1.1f%%', startangle=140, colors=plt.get_cmap('Set2').colors)
plt.title('Position Distribution')
plt.show()
```

```python
corr_matrix = df[['Age', 'Salary']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap: Age and Salary')
plt.show()
```
